## 네이버증권 뉴스 제목과 URL 수집하기

```python
https://finance.naver.com/
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
PATH = "https://finance.naver.com/"
resp = requests.get(PATH)
```

```python
resp
```

```python
resp?
```

```python
requests.get?
```

```python
resp.text
```

```python
print(resp.text)
```

```python
src = resp.text
```

```python
soup = BeautifulSoup(src, 'lxml')
print(soup)
```

```python
# 뉴스 목록 찾기
srclist = soup.select('.section_strategy li a')
srclist
```

```python
# 첫번째 뉴스 데이터 추출
srclist[0].text
```

```python
srclist[0]['href']
```

```python
url = srclist[0]['href']
PATH + url
```

```python
from urllib.parse import urljoin
```

```python
urljoin(PATH, url)
```

```python
news_title= []
news_url = []

for n in srclist:
    title=n.text
    url=urljoin(PATH, n['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)    
```

## 데이터 프레임 변환

```python
import pandas as pd
```

```python
df=pd.DataFrame({'제목':news_title, '주소':news_url})
df
```

## excel 파일 저장



```python
df.to_excel('output/naver.xlsx', index = False)
```
