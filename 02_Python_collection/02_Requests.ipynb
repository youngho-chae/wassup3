## 네이버증권 뉴스 제목과 URL 수집하기

```python
https://finance.naver.com/
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
PATH = "https://finance.naver.com/"
resp = requests.get(PATH)
```

```python
resp
```

```python
resp?
```

```python
requests.get?
```

```python
resp.text
```

```python
print(resp.text)
```

```python
src = resp.text
```

```python
soup = BeautifulSoup(src, 'lxml')
print(soup)
```

```python
# 뉴스 목록 찾기
srclist = soup.select('.section_strategy li a')
srclist
```

```python
# 첫번째 뉴스 데이터 추출
srclist[0].text
```

```python
srclist[0]['href']
```

```python
url = srclist[0]['href']
PATH + url
```

```python
from urllib.parse import urljoin
```

```python
urljoin(PATH, url)
```

```python
news_title= []
news_url = []

for n in srclist:
    title=n.text
    url=urljoin(PATH, n['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)    
```

## 데이터 프레임 변환

```python
import pandas as pd
```

```python
df=pd.DataFrame({'제목':news_title, '주소':news_url})
df
```

## excel 파일 저장



```python
df.to_excel('output/naver.xlsx', index = False)
```

## 파일명에 저장날짜 추가하기

```python
# 파일명에 저장시각을 추가해보자
import time
```

```python
today = time.localtime()
today
```

```python
df.to_excel(f'output/{today.tm_year}{today.tm_mon}{today.tm_mday}_naver.xlsx', index=False)
```

```python
# 스트링포매팅으로 날짜포맷을 바꿔보자
# %02d        % /   표현숫자 /   전체자리수 2자리 / 정수 
'%d-%02d-%02d' % (today.tm_year, today.tm_mon, today.tm_mday)
```

```python
file_name = '%d-%02d-%02d' % (today.tm_year, today.tm_mon, today.tm_mday)
excel_name = file_name + '.xlsx'
csv_name = file_name + '.csv'
print(excel_name, csv_name)
```

```python
df.to_excel(excel_name, index=False)
```

## 멜론차트, 노래제목과 가수 수집하기

```python
https://www.melon.com/chart/index.htm
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
PATH = 'https://www.melon.com/chart/index.htm'
resp = requests.get(PATH)
resp
```

```python
info = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'}
info
```

```python
requests.get?
```

```python
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
```

```python
soup = BeautifulSoup(html_src, 'lxml')
soup
```

```python
song = soup.select('.wrap span a')[0].text
song
```

```python
len(soup.select('.ellipsis.rank02 > a'))
```

```python
artist = []
for item in soup.select('.checkEllipsis'):
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
len(soup.select('.checkEllipsis'))
```

```python
name = [item.text for item in soup.select('.checkEllipsis')]
print(len(name))
print(name)
```

```python
rank = list((range(1, 101)))
rank
```

```python
# 판다스로 데이터 프레임 만들기 clumn : data
songDF = pd.DataFrame({'순위': rank,
             '노래제목': song,
             '가수명': name})
songDF
```

```python
songDF.to_excel('output/melon100.xlsx', index = False)
```

## REST test get/post 비교

https://blog.naver.com/gikimirane/222666913565

```python
import requests
```

```python
param = {'name':"Amy", 'age':20, 'address':'Seoul'}
```

```python
# get 방식
resp1 = requests.get('http://httpbin.org/get', params=param)
print(resp1)
print(resp1.text)
```

```python
#post 방식
resp2 = requests.get('http://httpbin.org/post', params=param)
print(resp2)
print(resp2.text)
```

```python
resp2 = requests.post('http://httpbin.org/post', data=param)
print(resp2)
print(resp2.text)
```
